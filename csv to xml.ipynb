```python
import pandas as pd
from google.colab import  files 
uploaded = files.upload()
```

```python
with open("data.csv", 'r') as temp_f:
    # get No of columns in each line
    col_count = [ len(l.split(";")) for l in temp_f.readlines() ]

### Generate column names  (names will be 0, 1, 2, ..., maximum columns - 1)
column_names = [i for i in range(0, max(col_count))]

df = pd.read_csv("data.csv", header=None, delimiter=";", names=column_names)
df = df.astype(object).where(pd.notnull(df),0)
df.head(25)
```

```python
df.shape
```

```python
def convert_row(row):
    return """<einsatz="%s">
    <plz>%s</plz>
    <name>%s</name>
    <strasse>%s</strasse>
    <kontakt><mobil>%s</mobil></kontakt>
    </einsatz>""" % (row[0], row[1], row[2], row[4],row[10])

xr = '\n'.join(df.apply(convert_row, axis=1))
```

```python
with open('example.txt', 'w') as f:
  f.write(xr)
files.download('example.txt')
```
